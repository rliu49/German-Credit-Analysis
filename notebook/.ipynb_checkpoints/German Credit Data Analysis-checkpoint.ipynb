```python
GermanCredit['Saving accounts'] = GermanCredit['Saving accounts'].map({"little":0,"moderate":1,"quite rich":2,"rich":3});
GermanCredit['Saving accounts'] = GermanCredit['Saving accounts'].fillna(GermanCredit['Saving accounts'].dropna().mean())

GermanCredit['Checking account'] = GermanCredit['Checking account'].map({"little":0,"moderate":1,"rich":2});
GermanCredit['Checking account'] = GermanCredit['Checking account'].fillna(GermanCredit['Checking account'].dropna().mean())
```

```python
GermanCredit['Sex'] = GermanCredit['Sex'].map({"male":0,"female":1}).astype(float)
GermanCredit['Housing'] = GermanCredit['Housing'].map({"own":0,"free":1,"rent":2}).astype(float)
GermanCredit['Purpose'] = GermanCredit['Purpose'].map({'radio/TV':0, 'education':1, 'furniture/equipment':2, 'car':3, 'business':4,
       'domestic appliances':5, 'repairs':6, 'vacation/others':7}).astype(float);

GermanCredit = GermanCredit.dropna(how='any')

GermanCredit.head()
```

```python
from sklearn.cluster import KMeans;
from sklearn.decomposition import PCA;
from sklearn.preprocessing import normalize;
```

```python
y = KMeans().fit_predict(GermanCredit)
X_norm = normalize(GermanCredit);
y_PCA = PCA(n_components=2).fit_transform(X_norm,2);
y_PCA.shape
```

```python
plt.scatter(GermanCredit['Credit amount'],GermanCredit['Age'],c=y)
plt.figure()
```

```python
plt.scatter(y_PCA[:,0],y_PCA[:,1],c=y)
```
